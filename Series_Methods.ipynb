Neste notebook, vamos explorar os principais métodos do objeto Series da biblioteca pandas, que é uma estrutura fundamental para análise de dados em Python. Uma Series pode ser entendida como uma coluna de dados indexada, semelhante a uma lista, porém muito mais poderosa por conter rótulos (índices) e diversos métodos prontos para manipulação e análise.

Os métodos de Series permitem realizar operações como estatísticas, limpeza de dados, transformação e exploração de informações de forma rápida e eficiente. Por exemplo, podemos calcular a média dos valores com mean(), somar elementos com sum(), visualizar os primeiros registros com head(), contar ocorrências com value_counts() e tratar valores ausentes com fillna().

##  Importando a bilioteca Pandas.

```python
import pandas as pd
```

## Criando a lista de Strings.


```python
nomes = ['Marcio','Dayane','Alice','Alana','Samuel']
```

## Criando um dicionário de numeros.


```python
idades = {
          'Marcio':42,
          'Dayane':40,
          'Alice':9,
          'Alana':7,
          'Samuel':1
         }
```

## Criando as Series.

```python
# Cria a serie (serie_nomes) a partir da lista de nomes.
serie_nomes = pd.Series(nomes)


#  Cria a serie (serie_idades) a partir do dicionário de idades.
serie_idades = pd.Series(idades)
```

## <u>SUM()<u> -  Soma todos os valores da Série


```python
# O pandas vai somar os valores, se todos forem int ele retorna int se tiver 1 float ele retorna float.
soma_idades = serie_idades.sum()
print(f'A soma das idades da fámilia é:{soma_idades} anos!')
```

## MEAN() - Média dos valores (Soma total / Quantidade)

```python
media_idades = serie_idades.mean()
print(f'A média de idades da familia é:{media_idades} anos!')
```

## MEDIAN() - Mediana dos valores(O valor central dos dados.)

> 
Exemplo:
Conjunto: 2, 4, 7, 9, 10
A mediana é 7 (o valor do meio).




```python
# A Alice é a mediana (Dois mais novos Dois mais velhos)
mediana_idade = serie_idades.median()
print(f'A mediana das idades da familia é:{mediana_idade}')
```

##  MODE() - O valor da moda (O valor que mais aparece.)

```python
''' Exemplos 
        P/ lista  [1,2,3,4,5]
        A moda é [1,2,3,4,5]  (todos os valores pq aparecem uma vez.)
        
        P/ a lista [1,1,2,3,4,5]
        A moda é 1. (O numero que mais aparece: apareceu duas vezes!)

        P/ a lista [1,1,2,2,3,4,5,6]
        A moda é 1,2 (Ambos numeros aparecem duas vezes!)
'''    
moda_idade = serie_idades.mode()
print(moda_idade)
```

## MIN() -  O menor valor dentro da serie

```python
# Serie string    = ordem alfabetica
# Serie numereica =  menor valor numérico
# ** Não esqueca! -12 é menor que -11 
# Em series em que o elementos são de tipos diferente o min não funciona.


valor_minimo = serie_idades.min()
print(f'O menor valor na serie é:{valor_minimo}, conserquentemente essa é a idade da pessoa mais nova!')

```

## MAX() -  O Maior valor dentro da serie

```python
# Serie string    = ordem alfabetica
# Serie numereica =  maior para maior
# Em series em que o elementos são de tipos diferente o min não funciona.

valor_maximo = serie_idades.max()
print(f'O valor máximo na serie é:{valor_maximo},consequentemente essa é a idade da pessoa mais velha!')



```

## COUNT() - Conta quantas linhas (Não nulas) tem na série 

```python
# Note que mesmo que a serie contenha valores numéricos o count da a quantidade de item e nao a soma deles!
quantidade_pessoas = serie_idades.count()
print(f'A quantidade de itens na serie é:{quantidade_pessoas}, consequentemente essa é a quantidade de membros da fámilia!')
```

## STD() - Desvio Padrão (Quão longe os valores estão da média.)

> Maior o desvio mais longe  da média os numeros estão.
> Menor o desvio mais perto da media os numeros estão.

```python
# **  Podemos aplicar o std somente em series numericas!
desvio_padrao = serie_idades.std()
print(f'O desvio padrão das idade é:{desvio_padrao}')
```

## VAR() - Variancia (A variância mede o quanto os valores de um conjunto estão dispersos em relação à média.)


```python
variancia = serie_idades.var()
print(f'A variancia das idades é:{variancia}')
```

## UNIQUE() - (Os valores nao repetidos da serie)

```python
# Quando houver repetição o primeiro a aparecer será mostrado.
unique = serie_idades.unique()
print(f'Os valores nao repetidos da serie sao: {unique}')
```

## NUNIQUE - (Quantidade de itens únicos na serie)

```python
nunique = serie_idades.nunique()
print(f'A quantidade de valores unicos na série é:{nunique}')
```

## VALUE_COUNTS - (Quantas vezes cada valor aparece na coluna ou serie)

```python
value_counts = serie_idades.value_counts()
print(f'A contagem de  quantas vezes cada valor aparece na séries é:{value_counts}')
```

## DESCRIBE - (Resumo estatistico da series)

```python
describe  = serie_idades.describe()
print(f'O resumo estatistico da serie é:{describe}')
```

## IDMAX(Indice do maior valor)

```python
idmax = serie_idades.idxmax()
print(f'O indice do maior valor é:{idmax}')
print(type(idmax))
```
