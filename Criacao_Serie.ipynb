# Criação de Series no Pandas

## O que são Series?

As `Series` são estruturas de dados do Pandas compostas por uma única dimensão (uma coluna), capazes de armazenar valores de um mesmo tipo de dado.  

Elas funcionam de forma semelhante a uma lista ou coluna de uma tabela, mas possuem um índice associado a cada elemento, o que facilita a manipulação e a consulta dos dados.

Neste tópico, aprenderemos como criar e utilizar `Series` no Pandas.


---

# Importando a biblioteca.

```python
# ** Uma boa prática é sempre adicionar um apelido para a biblioteca. (A comunidade usa pd para o Pandas)
import pandas as pd
```

# Criando uma lista para ser nossa "base de dados"

```python
# Isso cria a lista que serão os dados que vamos trabalhar.
sorvetes = ['Chocolate','Morango','Napolitano','Flocos','Limao','Uva']
```

# Criando nossos objeto Series.

### <u>A partir de uma lista.</u>

```python
# Abaixo criaremos uma Serie do tipo string contendo os dados da variavel( do tipo lista)  com  os sorvetes.
pd.Series(sorvetes)
```

```python
# Agora temos uma lista(var) com conteúdo de tipos diferentes (numero,string,bolean) portando a série criada é do tipo object.
var = [3,'Hello World',True]
pd.Series(var)
```

### <u>A partir de um dicionário</u>.


```python
# Criando o dicionário
familia = {
           'Papai':'Marcio',
           'Mamãe':'Dayane',
           'Filha_1':'Alice',
           'Filha_2':'Alana',
           'Filho':'Samuel',
          }
```

```python
# Note que agora teremos a chave do dicionário como índice da série.
pd.Series(familia)
```
