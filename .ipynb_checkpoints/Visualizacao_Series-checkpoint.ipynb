##  Importa a biblioteca e cria as Series

```python
# Importa a biblioteca
import pandas as pd
```

```python
#  O metodo squeeze está transformanto do  retorno  do read CSV que  queria um dataframe em uma Serie.
pokemon = pd.read_csv("/Users/user/Documents/GitHub/Python_Dados/data-analysis-with-pandas-and-python/Incomplete/pokemon.csv", usecols=["Name"]).squeeze("columns")
pokemon
```

```python
google = pd.read_csv("/Users/user/Documents/GitHub/Python_Dados/data-analysis-with-pandas-and-python/Incomplete/google_stock_prices.csv", usecols=["Price"]).squeeze("columns")
google
```

## LEN() - Nos dá os tamanho da Serie.

```python
# O Len funciona exatamente igual ao Python ou qualquer outra  linguagem.
len(pokemon)
```

## HEAD(X) - Exibe as "X" linhas de uma Serie


```python
#  Se você não passr numero nenhum irá retornar 5 linhas.
pokemon.head(4)
```

## TAIL(Y) -  Exibe as "Y" linhas de uma serie

```python
#  Se você não passr numero nenhum irá retornar 5 linhas
pokemon.tail(4)
```

## [] - Exibe  um elemento específico pelo Indice


```python
# Podemos também o fatiamento de String do Python.
#  Quando o indice for string  tem  que colocar ""  exemplo   pokemon.iloc["indice_0":"indice_1"]
pokemon[3]
pokemon[2:6]
```

## ILOC[] - Exibe um elemento especifico pela posição numérica do indice.

```python
# Como o Iloc pega a posição ele aceita somente numerobs.
pokemon.iloc[0]
```

## LOC[] - Exibe um elemento específico pelo nome do Indice.
    

```python
pokemon.loc[0]
```

## UNIQUE - Exibe os valores Unicos

```python
# Exibe os valore únicos de uma serie.
pokemon.unique()
```

## VALUE_COUNTS() - Mostra quantas vezes cada item aparece.

```python
pokemon.value_counts()
```

## SORT_VALUES() - Ordena os valores (Key = função customizada para ordenação.)

```python
pokemon.sort_index()
```

## serie[serie=="valor_a_ser_pesquisado"]

```python
# O código abaixo tenta encontrar o pokemon chamado Picachu
## **  A procura é case sensitive e  somente o conteúdo completo pode ser buscado dessa forma.
pokemon[pokemon=="Pikachu"]
```

##  STR.STARTSWITH - Tods items que comecem com "XX"  

```python
# Lembrando que o comando  Starts with  é case sensitive.
pokemon[pokemon.str.startswith('Pik')]
```

## STR.ENDSWITH - Todos os items que terminam com 'XX'

```python
# Lembrando que o endswith é case sensitive.
pokemon[pokemon.str.endswith('mar') ] 
```

## STR.CONTAINS - Todos os items que contém  algo


```python
pokemon[pokemon.str.contains('Pik', case=False)]
```

## TOLIST() - Transforma a Serie em uma lista

```python
lista = pokemon.to_list()
type(lista)
```

## SAMPLE(X) - Nos dá X exemplos aleatórios.    

```python
# Caso você não passe X será mostrado um único exemplo.
pokemon.sample(3)
```

## ISNULL() -  Nos dá se existem  valores nulos na série.


```python
# Esse metodo retorna uma serie contendo todos os indices e FALSE/TRUE na frente do Indice  dizer se tem ou nao  conteúdo naquele indice
pokemon.isnull()
```
