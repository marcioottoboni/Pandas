Neste notebook, vamos explorar os principais atributos do objeto Series da biblioteca pandas
Os atributos de uma Series fornecem informações importantes sobre sua estrutura, conteúdo e organização dos dados. Por exemplo, podemos visualizar os índices com index, identificar o tipo dos dados armazenados com dtype, acessar os valores da Series com values, verificar o tamanho utilizando size e analisar o formato através de shape. Esses atributos são extremamente úteis para compreender a estrutura dos dados antes de iniciar operações de análise, limpeza ou transformação.

## Importa a biblioteca e cria as series.

```python
#  Importa a biblioteca.
import pandas as pd
```

```python
# Cria as variáveis.
nomes = ['Marcio','Dayane','Alice','Alana','Samuel']


idades = {
          'Marcio':42,
          'Dayane':40,
          'Alice':9,
          'Alana':7,
          'Samuel':1
         }
```

```python
# Cria a serie (serie_nomes) a partir da lista de nomes.
serie_nomes = pd.Series(nomes)

#  Cria a serie (serie_idades) a partir do dicionário de idades.
serie_idades = pd.Series(idades)
```

## VALUES - Retorna os valoes de uma Series.

```python
values = serie_nomes.values
print(f'Os valores da Serie sao: {values}')
```

## INDEX - Retorna os índices da Serie.

```python
index = serie_idades.index
print(f'Os indices  da serie sao: {index}')
```

## VALUES - Retorna os valores da serie.

```python
values = serie_nomes.values
print(f'Os valores da serie sao: {values}')
```
